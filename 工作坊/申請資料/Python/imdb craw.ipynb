## 從IMDd爬取奧斯卡15年來的得獎片單（最佳影片、原創劇本、改編劇本、紀錄片）

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.imdb.com/event/ev0000003/2018/1/?ref_=ev_eh'
results = requests.get(url)
soup = BeautifulSoup(results.text, 'html.parser')
```

```python
list_a=[]
for n in range(2004,2019):
    url = 'https://www.imdb.com/event/ev0000003/{}/1/?ref_=ev_eh'.format(n)
    results = requests.get(url)
    soup = BeautifulSoup(results.text, 'html.parser')

    list_1=[]
    for div in soup.find_all(class_="article"):
        list_1.append(div.find(class_="ab_widget"))
    string_1=list_1[0].text.strip()
    list_2=string_1.split("\n")
    string_2=list_2[2].strip("]);")
    list_3=string_2.split("',")
    string_3=list_3[1]
    list_a.append(string_3)
print(len(list_a))
```

```python
import json
list_b=[]
for n in range(0,15):
    list_b.append(json.loads(list_a[n]))
#list_b[0]為dict_2004;list_b[1]為dict_2005...list_b[14]為dict_2018
```

```python
#觀察bestpicture
for n in range(0,15):
    print('---'+str(2004+n)+'---')
    print(list_b[n]['nomineesWidgetModel']['eventEditionSummary']['awards'][0]['categories'][0]['categoryName'])
    print('----------')
    for n2 in range(len(list_b[n]['nomineesWidgetModel']['eventEditionSummary']['awards'][0]['categories'][0]['nominations'])):
        print(list_b[n]['nomineesWidgetModel']['eventEditionSummary']['awards'][0]['categories'][0]['nominations'][n2]['primaryNominees'][0]['name'])
        print('https://www.imdb.com/title/' + list_b[n]['nomineesWidgetModel']['eventEditionSummary']['awards'][0]['categories'][0]['nominations'][n2]['primaryNominees'][0]['const'] )
```

```python
#製作bestpicture的15年電影名稱與網址清單
list_bestpicture=[]
for n in range(0,15):
    list_bestpicture.append(2004+n)
    list_bestpicture.append(0)
    list_bestpicture[(n+1)*2-1]=[]
    for n2 in range(len(list_b[n]['nomineesWidgetModel']['eventEditionSummary']['awards'][0]['categories'][0]['nominations'])):
        list_bestpicture[(n+1)*2-1].append(list_b[n]['nomineesWidgetModel']['eventEditionSummary']['awards'][0]['categories'][0]['nominations'][n2]['primaryNominees'][0]['name'])
print(list_bestpicture)
    
i=2
bestpicture_winner_movielist=[]
for m in list_bestpicture:
    if i/2==1:
        bestpicture_winner_movielist.append(m)
        i=0
    else:  
        bestpicture_winner_movielist.append(m[0])
    i+=1
print(bestpicture_winner_movielist)
#list_bestpicture架構為：[年份,[電影名稱，電影名稱...],年份,[電影名稱，電影名稱...]...]
```

```python
#儲存
with open('bestpicture_movielist.json','w') as outfile:
    json.dump(list_bestpicture,outfile)
with open('bestpicture_winner_movielist.json','w') as outfile:
    json.dump(bestpicture_winner_movielist,outfile)
```
