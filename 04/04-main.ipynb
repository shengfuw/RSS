```python
import random
import numpy as np
import math

class agent: 
    def __init__(self):
        self.G1, self.G2, self.G3, self.G4, self.G5, self.G6, self.G7, self.G8, self.G9, self.G10, self.G11, self.G12, self.G13, self.G14, self.G15 = random.choices([True, False], k=15)
        self.payoff_record = []

    def get_marker(self):
        if self.G2 == False:
            # agent is D 
            if self.G1 == False: return True
            # agent is C
            else: return False
        if self.G2 == True:
            # agent is C
            if self.G1 == True: return True
            # agent is D
            else: return False
    
    def get_fitness(self):
        n = len(self.payoff_record)
        if n == 0: return 0
        fitness_vector = np.array(self.payoff_record)
        weights = np.logspace(n-1, 0, n, base=.999)
        return fitness_vector.dot(weights) / np.sum(weights)
    
    def update_payoff(self, payoff, t):
        if len(self.payoff_record) < t: self.payoff_record.append(payoff)
        else: self.payoff_record[-1] += payoff

    def decide_turst(self, another_agent, relation_stranger):
        # Gene 14: Unconditional rule
        if self.G4 == False and self.G5 == False and self.G6 == False and self.G7 == False and self.G8 == False:
            return(self.G14) # G14 = 1 then return True (trust); G14 = 0 then return False (exit)

        trust_counter = 0
        distrust_counter = 0

        # gene 4, 9: Base trust on your own intentions
        if self.G4 == True:
            if self.G9 == True: trust_counter += 1
            else: distrust_counter += 1
        # gene 5, 10: Attend to partner's G1 marker
        if self.G5 == True:
            if self.G10 == True:
                if another_agent.get_marker() == True: trust_counter += 1  
                else: distrust_counter += 1
            else:
                if another_agent.get_marker() == False: trust_counter += 1  
                else: distrust_counter += 1
        # gene 6, 11: Attend to partner's greeting
        if self.G6 == True:
            if self.G11 == True:
                if another_agent.G3 == True: trust_counter += 1  
                else: distrust_counter += 1
            else:
                if another_agent.G3 == False: trust_counter += 1  
                else: distrust_counter += 1
        # gene 7, 12: Attend to partner's group membership
        if self.G7 == True:
            if self.G12 == True:
                if relation_stranger == False: trust_counter += 1
                else: distrust_counter += 1
            else:
                if relation_stranger == True: trust_counter += 1
                else: distrust_counter += 1
        # gene 8, 13: Attend to partner's relative successif self.G4 == True:
        if self.G8 == True:
            if self.G13 == True:
                if another_agent.get_fitness() > self.get_fitness(): trust_counter += 1
                else: distrust_counter += 1
            else:
                if another_agent.get_fitness() <= self.get_fitness(): trust_counter += 1
                else: distrust_counter += 1
        
        information = []
        for i in range(trust_counter):
            information.append(True)
        for i in range(distrust_counter):
            information.append(False)
        return random.choice(information)

    def decide_cooperate(self, relation_stranger):
        if self.G1 == True: # agent is C
            return True
        else: # agent is D
            if self.G15 == True:
                if relation_stranger == True: return False
                else: return True
            else:
                return False

def transfer_chromosome(higher_payoff_agent, less_payoff_agent):
    # print("*", higher_payoff_agent.G1, higher_payoff_agent.get_fitness(), less_payoff_agent.G1, less_payoff_agent.get_fitness())

    if random.choice([True, False]): 
        # print(less_payoff_agent.G1, higher_payoff_agent.G1)
        if random.randint(1,100) == 1: less_payoff_agent.G1 = not higher_payoff_agent.G1
        else: less_payoff_agent.G1 = higher_payoff_agent.G1
        # print(less_payoff_agent.G1, higher_payoff_agent.G1)
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G2 = not higher_payoff_agent.G2
        else: less_payoff_agent.G2 = higher_payoff_agent.G2
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G3 = not higher_payoff_agent.G3
        else: less_payoff_agent.G3 = higher_payoff_agent.G3
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G4 = not higher_payoff_agent.G4
        else: less_payoff_agent.G4 = higher_payoff_agent.G4
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G5 = not higher_payoff_agent.G5
        else: less_payoff_agent.G5 = higher_payoff_agent.G5
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G6 = not higher_payoff_agent.G6
        else: less_payoff_agent.G6 = higher_payoff_agent.G6
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G7 = not higher_payoff_agent.G7
        else: less_payoff_agent.G7 = higher_payoff_agent.G7
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G8 = not higher_payoff_agent.G8
        else: less_payoff_agent.G8 = higher_payoff_agent.G8
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G9 = not higher_payoff_agent.G9
        else: less_payoff_agent.G9 = higher_payoff_agent.G9
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G10 = not higher_payoff_agent.G10
        else: less_payoff_agent.G10 = higher_payoff_agent.G10
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G11 = not higher_payoff_agent.G11
        else: less_payoff_agent.G11 = higher_payoff_agent.G11
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G12 = not higher_payoff_agent.G12
        else: less_payoff_agent.G12 = higher_payoff_agent.G12
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G13 = not higher_payoff_agent.G13
        else: less_payoff_agent.G13 = higher_payoff_agent.G13
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G14 = not higher_payoff_agent.G14
        else: less_payoff_agent.G14 = higher_payoff_agent.G14
    if random.choice([True, False]):
        if random.randint(1,100) == 1: less_payoff_agent.G15 = not higher_payoff_agent.G15
        else: less_payoff_agent.G15 = higher_payoff_agent.G15

def encounter(a_agent, b_agent, relation_stranger, t, records) -> dict:
    global X, R, S, T, P

    a_trust = a_agent.decide_turst(b_agent, relation_stranger)
    b_trust = b_agent.decide_turst(a_agent, relation_stranger)

    if a_trust and b_trust:
        if relation_stranger: records["Trust in strangers"][-1] += 2
        else: records["Trust in neighbors"][-1] += 2

        if a_agent.decide_cooperate(relation_stranger) == True and b_agent.decide_cooperate(relation_stranger) == True:
            a_agent.update_payoff(R, t)
            b_agent.update_payoff(R, t)
            if relation_stranger: records["Cooperation with strangers"][-1] += 2
            else: records["Cooperation with neighbors"][-1] += 2

        elif a_agent.decide_cooperate(relation_stranger) == True and b_agent.decide_cooperate(relation_stranger) == False:
            a_agent.update_payoff(S, t)
            b_agent.update_payoff(T, t)
            if relation_stranger: records["Cooperation with strangers"][-1] += 1
            else: records["Cooperation with neighbors"][-1] += 1

        elif a_agent.decide_cooperate(relation_stranger) == False and b_agent.decide_cooperate(relation_stranger) == True:
            a_agent.update_payoff(T, t)
            b_agent.update_payoff(S, t)
            if relation_stranger: records["Cooperation with strangers"][-1] += 1
            else: records["Cooperation with neighbors"][-1] += 1

        elif a_agent.decide_cooperate(relation_stranger) == False and b_agent.decide_cooperate(relation_stranger) == False:
            a_agent.update_payoff(P, t)
            b_agent.update_payoff(P, t)

        a_fitness = a_agent.get_fitness()
        b_fitness = b_agent.get_fitness()
        if a_fitness > b_fitness:
            transfer_chromosome(a_agent, b_agent)
        elif a_fitness < b_fitness:
            transfer_chromosome(b_agent, a_agent)       
            
    else:
        # if relation_stranger: records["Trust in strangers"][-1] += a_trust + b_trust
        # else: records["Trust in neighbors"][-1] += a_trust + b_trust
        a_agent.update_payoff(X, t)
        b_agent.update_payoff(X, t)
        
    return records

class lattice:
    def __init__(self, agent_num:int, neighborhood_size:int, embeddedness, get_record=False):
        self.agents = []
        for i in range(agent_num):
            self.agents.append(agent())
        self.size = agent_num 
        self.S = neighborhood_size
        self.Z = math.ceil(agent_num/neighborhood_size)
        self.embeddedness = embeddedness
        self.round_num = 0
        self.get_record = get_record
        self.records = {"Iteration":[], "G1":[], "G2":[], "G3":[], "G4":[], "G5":[], "G6":[], "G7":[], "G8":[], "G9":[], "G10":[], "G11":[], "G12":[], "G13":[], "G14":[], "G15":[], "Pairied with neighbor":[], "Pairied with stranger":[], "Trust in neighbors":[], "Trust in strangers":[], "Cooperation with neighbors":[], "Cooperation with strangers":[]}

    def one_round(self):
        self.round_num += 1
        self.records["Iteration"].append(self.round_num)
        self.records["Pairied with neighbor"].append(0)
        self.records["Pairied with stranger"].append(0)
        self.records["Trust in neighbors"].append(0)
        self.records["Trust in strangers"].append(0)
        self.records["Cooperation with neighbors"].append(0)
        self.records["Cooperation with strangers"].append(0)

        selected_agents_idx = []
        for i in range(1, self.size+1):
            if i not in selected_agents_idx:
                zi = math.ceil(i/self.S)
                if random.uniform(0, 1) <= self.embeddedness: # assign a partner from neighborhood
                    partner_idx = (i - 2 + random.randint(2,self.S)) % self.S + 1 + (zi-1)*self.S
                    self.records = encounter(self.agents[i-1], self.agents[partner_idx-1], False, self.round_num, self.records) # because list index start from 0, so need minus 1     
                    self.records["Pairied with neighbor"][-1] += 1 
                    # print(i, partner_idx, "N")              
                else: # assign a stranger
                    partner_idx = random.randint(1,self.S) + ((zi - 2 + random.randint(2,self.Z)) % self.Z) * self.S
                    self.records = encounter(self.agents[i-1], self.agents[partner_idx-1], True, self.round_num, self.records)
                    self.records["Pairied with stranger"][-1] += 1 
                    # print(i, partner_idx, "S")    
 
                selected_agents_idx.append(partner_idx)

        # for i in range(self.size):
        #     print(self.agents[i].G1, self.agents[i].payoff_record)

        if self.get_record:
            genes = [(agent.G1, agent.G2, agent.G3, agent.G4, agent.G5, agent.G6, agent.G7, agent.G8, agent.G9, agent.G10, agent.G11, agent.G12, agent.G13, agent.G14, agent.G15) for agent in self.agents]
            genes_ratio = list(np.array(genes).mean(axis=0))
            self.records["G1"].append(genes_ratio[0])
            self.records["G2"].append(genes_ratio[1])
            self.records["G3"].append(genes_ratio[2])
            self.records["G4"].append(genes_ratio[3])
            self.records["G5"].append(genes_ratio[4])
            self.records["G6"].append(genes_ratio[5])
            self.records["G7"].append(genes_ratio[6])
            self.records["G8"].append(genes_ratio[7])
            self.records["G9"].append(genes_ratio[8])
            self.records["G10"].append(genes_ratio[9])
            self.records["G11"].append(genes_ratio[10])
            self.records["G12"].append(genes_ratio[11])
            self.records["G13"].append(genes_ratio[12])
            self.records["G14"].append(genes_ratio[13])
            self.records["G15"].append(genes_ratio[14])
```

```python
# random.seed(24)
X = 2
R = 3
S = 0
T = 4
P = 1

Mylattice = lattice(1000, 10, 0.67, True)
for r in range(7000):
    if r % 500 == 0 and r != 0: print(r)
    Mylattice.one_round()

# print(Mylattice.records['G1'])
```
